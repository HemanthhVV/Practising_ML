```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/vvhem/Downloads/Sample_data1.csv.")
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Department=le.fit_transform(df.Department)
```

## df

```python
y=df.Placement
X=df.drop(['Placement','Name'],axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_traih,y_test=train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
%matplotlib inline
plt.plot(df.CGPA,df.Technical_Skills)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(X_train,y_traih)
```

```python
reg.score(X_test,y_test)
```

```python
reg.predict([[65,90,0,8.50,3,1,0,2]])
```

```python
reg.predict([[95,90,92,8.89,4,1,5,2]])
```
