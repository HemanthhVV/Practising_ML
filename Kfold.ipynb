```python
def model_ready(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return(model.score(X_test,y_test))
```

```python
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=3)
```

```python
for train_index_range,test_index_range in skf.split(df.data,df.target):
    X_tr,X_te,y_tr,y_te=df.data[train_index_range],df.data[test_index_range],df.target[train_index_range],df.target[test_index_range]
    print("Logistic Regression: ",model_ready(LogisticRegression(),X_tr,X_te,y_tr,y_te))
    print("Support Vector: ",model_ready(SVC(),X_tr,X_te,y_tr,y_te))
    print("Random Forest: ",model_ready(RandomForestClassifier(),X_tr,X_te,y_tr,y_te))
    print("\n")
```

```python
logis_reg=[]
svc=[]
forest=[]

for i in range(3,14):
    logis_reg.append(cross_val_score(LogisticRegression(),df.data,df.target,cv=i))
    svc.append(cross_val_score(LogisticRegression(),df.data,df.target,cv=i))
    forest.append(cross_val_score(LogisticRegression(),df.data,df.target,cv=i))
```

```python
import numpy as np
lo_av=[]
svc_av=[]
forest_av=[]
for i in logis_reg:
    lo_av.append(np.average(i))
for i in svc:
    svc_av.append(np.average(i))
for i in forest:
    forest_av.append(np.average(i))
```

```python
best=[np.average(lo_av),np.average(svc_av),np.average(forest_av)]
```
